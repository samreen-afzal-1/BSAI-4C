```python
# Water Jug Problem using DFS

def water_jug_dfs(cap1, cap2, goal):
    stack = [(0, 0, [])]   # (jug1, jug2, path)
    visited = set()

    while stack:
        jug1, jug2, path = stack.pop()

        if (jug1, jug2) in visited:
            continue

        visited.add((jug1, jug2))

        # Print current state
        print(f"Current State: ({jug1}, {jug2})")

        if jug1 == goal or jug2 == goal:
            print("\n Goal Achieved!")
            for step in path:
                print(step)
            print(f"Final State: ({jug1}, {jug2})")
            return

        # All possible moves
        next_states = []

        # 1. Fill Jug1
        next_states.append((cap1, jug2, path + ["Fill Jug1"]))

        # 2. Fill Jug2
        next_states.append((jug1, cap2, path + ["Fill Jug2"]))

        # 3. Empty Jug1
        next_states.append((0, jug2, path + ["Empty Jug1"]))

        # 4. Empty Jug2
        next_states.append((jug1, 0, path + ["Empty Jug2"]))

        # 5. Pour Jug1 → Jug2
        transfer = min(jug1, cap2 - jug2)
        next_states.append((jug1 - transfer, jug2 + transfer, path + ["Pour Jug1 → Jug2"]))

        # 6. Pour Jug2 → Jug1
        transfer = min(jug2, cap1 - jug1)
        next_states.append((jug1 + transfer, jug2 - transfer, path + ["Pour Jug2 → Jug1"]))

        for state in next_states:
            stack.append(state)

    print(" No solution found!")


# Example
cap1 = int(input("Enter capacity of Jug1: "))
cap2 = int(input("Enter capacity of Jug2: "))
goal = int(input("Enter goal amount: "))

water_jug_dfs(cap1, cap2, goal)
```
